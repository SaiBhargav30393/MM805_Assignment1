## SIFT Feature Detection


Here I am taking a image with RGB scale and converting it in Grey scale and after converting to grey scaling I am detecting features using SIFT and displaying features, later point we are also taking the variations of the images by rotating and scaling and detecting features and displaying

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python


def feature_detector(rgb_image):
    # This is used to to convert coloured image to grey scale
    grey_scaled = cv2.cvtColor(rgb_image, cv2.COLOR_RGB2GRAY)

    # We can use Surf if sift for variation
    features = cv2.SIFT_create()

    # To get the keypoints and descriptors by applying detectAndCompute method
    keypoints, descriptors = features.detectAndCompute(grey_scaled, None)
    
    # Images after applying keypoints and descriptors
    FinImg = cv2.drawKeypoints(rgb_image, keypoints, None, flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
    
    # Returning resulting image
    return FinImg
```

```python
# Reading image
rgb_image = cv2.cvtColor(cv2.imread('C:\\Users\\saibh\\Documents\\adobe_panoramas\\data\\fishbowl\\fishbowl-04.png'), cv2.COLOR_BGR2RGB)
# Detectinf features
detected_image = feature_detector(rgb_image)
# Display image
figuresize = (10, 10)
plt.figure(figsize=figuresize)
plt.imshow(detected_image)
plt.title("Feature Image")
plt.show()

```

```python
### Feature detector after rotation
## Take the image and rotate it
rgb = cv2.rotate(rgb_image, cv2.cv2.ROTATE_90_CLOCKWISE)
rotated_image = feature_detector(rgb)
# Display image
figuresize = (10, 10)
plt.figure(figsize=figuresize)
plt.imshow(rotated_image)
plt.title("Rotated image")
plt.show()
```

```python
def resized_img(image,scaling):
    # Scaling
    w = int(image.shape[1] * scaling / 100)
    h = int(image.shape[0] * scaling / 100)
    dimension = (w, h)

    # Resizing image after scaling
    resized = cv2.resize(image, dimension, interpolation = cv2.INTER_AREA)
    return resized
```

```python
### Feature detector after Scaling
rgb_scaled = resized_img(rgb_image,150)
scaled_image = feature_detector(rgb_scaled)
# Display Image
figuresize = (10, 10)
plt.figure(figsize=figuresize)
plt.imshow(scaled_image)
plt.title("scaled image")
plt.show()
```
