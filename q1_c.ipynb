```python
# Importing the Library
import cv2 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Code Reference

# https://docs.opencv.org/3.4/d4/d8c/tutorial_py_shi_tomasi.html
```

```python
img_RGB = cv2.imread('C:\\Users\\saibh\\Documents\\adobe_panoramas\\data\\fishbowl\\fishbowl-00.png')
grey_scale = cv2.cvtColor(img_RGB,cv2.COLOR_BGR2GRAY)

crn = cv2.goodFeaturesToTrack(grey_scale,25,0.01,10)
crn = np.int0(crn)

plt.imshow(img_RGB)
for i in crn:
    a,b = i.ravel()
    plt.scatter(a, b, 70, c="b", marker="o")
plt.show()
```
